# Coronavirus Disease-19 Data Visualisation

### Coronavirus disease (COVID-19) is an infectious disease caused by a newly discovered coronavirus.
The first case was identified in Wuhan, China, in December 2019. The disease has since spread worldwide, leading to an ongoing pandemic.

### Here we have visualise the covid-19 cases in India since it was first reported.

#### Firstly we will import all the required libraries

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

#### Now import the dataset and analyse it using various methods

```python
df_covid=pd.read_csv("../Projects/Datasets/covid_19_india.csv")
```

```python
df_covid.head()
```

#### Grouping the dataset keeping state/unionterritory as primary column

```python
df_state=df_covid.groupby(df_covid['State/UnionTerritory'],as_index=False).max()
```

```python
df_state.drop(['Sno','Date','Time','ConfirmedIndianNational','ConfirmedForeignNational'],axis=1,inplace=True)
```

```python
df_state.drop([7,24,32,36,37,39,41],inplace=True)
```

```python
df_state
```

#### Visualising the confirmed cases, recovered cases and deaths

```python
'''Plot the State vs Total confirmed cases in India/Cured/Deaths'''
fig,ax=plt.subplots(figsize=[140,10])
plt.bar(df_state['State/UnionTerritory'],df_state['Confirmed'])
plt.xlabel('States/Union Territory')
plt.ylabel('Confirmed Cases')
plt.title('State v/s Confirmed Cases')
for index,value in enumerate(df_state['Confirmed']):
    plt.text(index,value,value,fontsize=15,horizontalalignment='center')
fig,ax=plt.subplots(figsize=[140,10])
plt.bar(df_state['State/UnionTerritory'],df_state['Cured'],color='orange')
plt.xlabel('States/Union Territory')
plt.ylabel('Cured Cases')
plt.title('State v/s Cured Cases')
for index,value in enumerate(df_state['Cured']):
    plt.text(index,value,value,fontsize=15,horizontalalignment='center')
fig,ax=plt.subplots(figsize=[140,10])
plt.bar(df_state['State/UnionTerritory'],df_state['Deaths'],color='green')
plt.xlabel('States/Union Territory')
plt.ylabel('Deaths')
plt.title('State v/s Deaths')
for index,value in enumerate(df_state['Deaths']):
    plt.text(index,value,value,fontsize=15,horizontalalignment='center')
```

#### Now visualising the cases with time

```python
df_covid['Date']=pd.to_datetime(df_covid['Date'],format='%d/%m/%y')
```

```python
fig,ax=plt.subplots(figsize=[140,10])
plt.bar(df_covid['Date'],df_covid['Confirmed'])
plt.xlabel('Date')
plt.ylabel('Confirmed Cases')
plt.title('Date v/s Confirmed Cases')
fig,ax=plt.subplots(figsize=[140,10])
plt.bar(df_covid['Date'],df_covid['Cured'],color='orange')
plt.xlabel('Date')
plt.ylabel('Cured Cases')
plt.title('Date v/s Cured Cases')
fig,ax=plt.subplots(figsize=[140,10])
plt.bar(df_covid['Date'],df_covid['Deaths'],color='green')
plt.xlabel('Date')
plt.ylabel('Deaths')
plt.title('Date v/s Deaths')
```

#### Grouping the data keeping date as primary column

```python
df_date=df_covid.groupby(df_covid['Date'],as_index=False).sum()
```

```python
df_date.head()
```

```python
df_date.tail()
```

```python
# Getting daily new corona cases
x=df_date.drop(['Date'],axis=1)
df_daily = x.diff(axis=0)
```

```python
df_daily['Date']=df_date['Date']
df_daily.tail()
```

#### Visualising the cases with date

```python
fig,ax=plt.subplots(figsize=[140,10])
plt.bar(df_daily['Date'],df_daily['Confirmed'])
plt.xlabel('Date')
plt.ylabel('Confirmed Cases')
plt.title('Date v/s Confirmed Cases')
fig,ax=plt.subplots(figsize=[140,10])
plt.bar(df_daily['Date'],df_daily['Cured'],color='orange')
plt.xlabel('Date')
plt.ylabel('Cured Cases')
plt.title('Date v/s Cured Cases')
fig,ax=plt.subplots(figsize=[140,10])
plt.bar(df_daily['Date'],df_daily['Deaths'],color='green')
plt.xlabel('Date')
plt.ylabel('Deaths')
plt.title('Date v/s Deaths')
```

```python
ax,fig=plt.subplots(figsize=[100,10])
plt.plot(df_daily['Date'],df_daily['Confirmed'],linewidth=7,label='Confirmed Cases')
plt.plot(df_daily['Date'],df_daily['Cured'],linewidth=7,label='Cured Cases')
plt.plot(df_daily['Date'],df_daily['Deaths'],linewidth=7,label='Death')
plt.legend()
```

#### Getting recovery rate, death rate and CFR

```python
df_daily['Total_recovery']=(((df_daily[['Cured','Confirmed']]).min(axis=1))/((df_daily[['Cured','Confirmed']]).max(axis=1)))*100
df_daily['Total_death']=(((df_daily[['Deaths','Confirmed']]).min(axis=1))/((df_daily[['Deaths','Confirmed']]).max(axis=1)))*100
df_daily['CFR']=(df_daily['Deaths']/df_daily['Cured'])*100
```

```python
df_daily.tail()
```

#### Visualising different rates with date

```python
with sns.axes_style(style='darkgrid'):
    ax,fig=plt.subplots(figsize=[100,10])
    plt.plot(df_daily['Date'],df_daily['Total_recovery'],linewidth=7,label='Total_recovery')
    plt.plot(df_daily['Date'],df_daily['Total_death'],linewidth=7,label='Total_death')
    plt.plot(df_daily['Date'],df_daily['CFR'],linewidth=7,label='CFR')
    plt.legend()
```
